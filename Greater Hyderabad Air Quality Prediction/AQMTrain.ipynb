```python
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size=0.3, random_state=0)
from sklearn.linear_model import LinearRegression
reg_model=LinearRegression(normalize=True)
reg_model.fit(X_train,y_train)
reg_model.score(X_train,y_train)
reg_model.score(X_test,y_test)
#reg_model.coef_array([-8.71256785, -0.78565823, -0.64082652,  2.64932719, -1.44568242,0.27712587, -1.83610819, -1.01908448])
# for intercept
reg_model.intercept_

# cross validation
from sklearn.model_selection import cross_val_score
score=cross_val_score(reg_model,df.iloc[:,:-1],df.iloc[:,-1],cv=5)
score.mean()

prediction=reg_model.predict(X_test)
# checking predicted y and labeled y
plt.scatter(y_test,prediction)
metrics.mean_squared_error(y_test, prediction)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import randint
parameters = {'n_estimators': randint(100,1200),
               'max_features': ['auto','sqrt'],
               'max_depth': randint(5,40),
               'min_samples_split': randint(2,30),
               'min_samples_leaf': randint(1,10)  }
# Model for tuning
base_learner=RandomForestRegressor()
tuned_model= RandomizedSearchCV(estimator = base_learner, param_distributions = parameters,scoring='neg_mean_squared_error', n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs =-1)
tuned_model.fit(X_train,y_train)
tuned_model.best_params_

tuned_model.best_score_

prediction=tuned_model.predict(X_test)

plt.scatter(y_test,prediction)

metrics.mean_squared_error(y_test, prediction)
```

Random forest is a Generalized model compare to KNN so I deploy Random Forest Regressor with the top 5 features using streamlit
Random Forest Regressor for Top 5 features

```python

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,[0,1,2,3,6]], df.iloc[:,-1], test_size=0.3, random_state=0)

random_forest_reg1=RandomForestRegressor(n_estimators=901, max_depth=5, min_samples_split=16, min_samples_leaf=1, max_features='sqrt', n_jobs=-1)

random_forest_reg1.fit(X_train,y_train)
random_forest_reg1.score(X_train,y_train)

random_forest_reg1.score(X_test,y_test)

prediction=random_forest_reg1.predict(X_test)
metrics.mean_squared_error(y_test, prediction)
```

```python
import numpy as np
import pandas as pd
import pickle
import streamlit as st



pickle_in = open("Random_forest_regressor.pkl","rb")
random_forest_regressor=pickle.load(pickle_in)


def welcome():
    return " welcome all"



def predict_AQI(Average_Temperature,Maximum_Temperature,Minimum_Temperature,Atm_pressure_at_sea_level,Average_wind_speed):
    
    
    prediction=random_forest_regressor.predict([[ Average_Temperature,Maximum_Temperature,Minimum_Temperature, Atm_pressure_at_sea_level,Average_wind_speed]])
    print(prediction)
    return prediction


def main():
    st.title("Hyderabad AQI prediction")
    html_temp = """
    <div style="background-color:green;padding:20px">
    <h2 style="color:white;text-align:center;">AQI prediction ML App </h2>
    </div>
    """
    st.markdown(html_temp,unsafe_allow_html=True)
    Average_Temperature= st.text_input("Average_Temperature ","Type Here")
    Maximum_Temperature = st.text_input("Maximum_Temperature ","Type Here")
    Minimum_Temperature = st.text_input("Minimum_Temperature ","Type Here")
    Atm_pressure_at_sea_level = st.text_input("Atm_pressure_at_sea_level ","Type Here")
    Average_wind_speed = st.text_input("Average_wind_speed ","Type Here")
    result=""
    if st.button("Predict"):
        result=predict_AQI(Average_Temperature,Maximum_Temperature,Minimum_Temperature,Atm_pressure_at_sea_level,Average_wind_speed)
    st.success('The output is {}'.format(result))
    if st.button("About"):
        st.text("Janibasha Shaik")
        st.text(" 2020 ")

if __name__=='__main__':
    main()
```
